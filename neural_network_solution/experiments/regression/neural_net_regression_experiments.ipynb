# Experiment with Regression Problem

Initially, we intented to create two seperte embedding networks, one for classification tasks and second one for regression tasks. While building our solution, we discovered that the results for the regression tasks framework are not good enough.
It might be that the network that we built is not good enough and perhaps different architecture for the network would work better (we tried two different architectures). Another assumption is that the embedding of catagorical columns worse with regression tasks.
We chose to not further research this point and to convert an input regression task to a classification task with predicting above/under mean.

This notebook presents an example run of the solution with the embedding network as a regression task.


```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import sys
import os
root = os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath(''))))
sys.path.insert(0, root) 
from neural_net_for_regression_task import main as regression_main
import datetime
```

```python
data_path = '../../../datasets/converted_datasets/video_games_sales_converted.csv'
data = pd.read_csv(data_path)
data.head()
```

- The target column to predict is the Global_Sales column. 



- We know that the following columns have less than 50 unique INT values - therefore, they are actually categorical: Year_of_Release, Platform and Genre.


- Additionaly, we know that Year_of_Release is ordinal, while all the other are nominal.

We apply our model on the spotify dataset:

```python
warnings.filterwarnings("ignore")
start = datetime.datetime.now()
regression_main(data_path, 'Global_Sales')
end = datetime.datetime.now()
print("\n Model run time: {}".format(end-start))
```

We notice that the 'Year_of_Release' column was mistakenly classified as nominal.

We will now show that with the classification framework and converting the regression task to classification of above/below mean, the results will be better. 

We will use the main code of our final solution, giving it a regression_task flag that will convert the task to classification.


```python
from main import main
```

```python
warnings.filterwarnings("ignore")
start = datetime.datetime.now()
main(data_path, 'Global_Sales', regression_task=True)
end = datetime.datetime.now()
print("\n Model run time: {}".format(end-start))
```
